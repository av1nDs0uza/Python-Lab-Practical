```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# load data from a CSV file into pandas dataframe
df=pd.read_csv('Automobile_data.csv')
```

```python
#display column names
df.columns
```

```python
# to display total count of values including all rows and columns, rows * columns
df.size
```

```python
# to display No. of rows, column names, datatypes,
df.info()
```

```python
# to display top 5 rows with all the columns
df.head()
```

```python
# to check for wrong entries like symbols -,?,#,*,etc. in each column
for col in df.columns:
 print('{} : {}'.format(col,df[col].unique()))
```

```python
# to display missing values, python doesnot treat ?, & as missing value, it shows no  
# missing values as there are no NaN values
df.isnull().sum()
```

```python
# To replace symbols such as '?' , '&', .... with np.nan which indicates missing values
# After replacing all symbols, the new version of df is stored in df_clean..
df_clean = df.replace({ "?": np.nan, "&": np.nan })
```

```python
# to display total count of missing values in each column e.g. normalized-losses is hav
# ing 41 missing values
df_clean.isnull().sum()
```

```python
# to visualize missing values, yellow color indicate missing values in normalized-losse
# s, and in other columns....
sns.heatmap(df_clean.isnull(),cbar=False,cmap='viridis')
```

```python
#to display average value of each column...
print('Column Name\t\tAvg')
df_clean.mean()
```

```python
#to display descriptive stats of each column such as min, max, etc...
df_clean.describe()
```

```python
#to display top 5 rows with all columns
df_clean.head()
```

```python
#Histogram to visualize the distribution of engine-size column values, the max frequenc
#y count can be seen within 80-100 range
df['engine-size'].hist(bins=50, grid=False, xlabelsize=12, ylabelsize=12, color="blue")
plt.title('Distri Engine Size', fontsize=18)
plt.xlabel('Engine size', fontsize=13)
plt.ylabel('Count', fontsize=13)
```

```python
# to display th outliers present is specified column, the plot shows there are a few ou
# tliers present as indicated by dots
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 15]})
ax2=sns.boxplot(data=df['engine-size'], palette="Set3")
#plt.ylim([40.0,3200.0])
plt.title('Price v/s Engine Size', fontsize=18)
plt.xlabel('Price', fontsize=13)
plt.ylabel('Engine-Size', fontsize=13)
```

```python
# To display correlation matrix, dark colors (values close to +1) indicates strong posi
#tive correlation, 
# whereas light colors show correlation in decreasing intensity or negative correlation
#between independent variables or 
# independent and dependent variables (price)
plt.figure(figsize=(10,10))
sns.heatmap(df_clean.corr(),cbar=True,annot=True,cmap='Blues')
```
